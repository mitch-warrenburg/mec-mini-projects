# Iris Flower Dataset
This uses a simple dataset to learn the basics. It contains 3 flowers of iris species. It was created in 1936 by Ronald Fisher.

Link: [Iris Dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set)


Example of a petal and sepal:

![Iris with Labels](https://terpconnect.umd.edu/~petersd/666/html/iris_with_labels.jpg)

#### Objectives
Here are the objectives of this notebook:
1. Classify a new flower as belonging to one of the 3 classes given the 4 features
2. Learn we use petal and sepal dimensions as features
3. Learn why we don't use color as a feature


```python
import pandas as pd
import matplotlib.pyplot as pyplot
import seaborn

data = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")

print(data.shape)
print(data.columns.values)
print(data["species"].value_counts())
```

```python
data.plot(kind='scatter', x='sepal_length', y='sepal_width')
pyplot.show()
```

```python
seaborn.FacetGrid(data, hue="species", height=4) \
    .map(pyplot.scatter, "sepal_length", "sepal_width") \
    .add_legend()

pyplot.show()
```

**Observation(s):**
1. Using sepal_length and sepal_width features, we can distinguish Setosa flowers from others.
2. Seperating Versicolor from Viginica is much harder as they have considerable overlap.
